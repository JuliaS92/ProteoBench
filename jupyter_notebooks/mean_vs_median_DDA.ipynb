```python
import pandas as pd
```

```python
mean_data = pd.read_csv('/mnt/d/proteobench_dda/mean_result_table.csv')
median_data = pd.read_csv('/mnt/d/proteobench_dda/median_results_table.csv')
```

```python
mean_data_newpoints = mean_data[mean_data['is_temporary'] == True]
median_data_newpoints = median_data[median_data['is_temporary'] == True]
```

```python
mean_data_newpoints.columns
```

```python
# Plot mean_abs_epsilon vs nr_prec
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(10, 10))
ax = sns.scatterplot(y='nr_prec', x='median_abs_epsilon', data=mean_data_newpoints, hue='software_name', palette='viridis', s=200)
plt.ylabel('Total number of precursor ions quantified in the selected number of raw files')
plt.xlabel('Mean absolute difference between measured and expected log2-transformed fold change')
```
